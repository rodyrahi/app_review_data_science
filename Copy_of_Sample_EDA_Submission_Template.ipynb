````python
# Chart - 4 visualization code

# distribution of rating 


def convert_to_mb(value):
    value = value.lower()

    if 'm' in value:
        return float(value.replace('m', ''))  # assuming 'm' denotes megabytes
    elif 'k' in value:
        return float(value.replace('k', '')) / 1024  # converting kilobytes to megabytes
    else:
        return None


# ps_cleaned_df["Size"].unique()
ps_cleaned_df["size_num"] = ps_cleaned_df["Size"].apply(convert_to_mb)
ps_cleaned_df.shape
````
